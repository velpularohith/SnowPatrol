```python
import pandas as pd
import numpy as np
from datetime import timedelta, datetime
```

```python

holiday_dates = [datetime(2023, 1, 1), datetime(2023, 1, 26)]
end_date = datetime(2023, 8, 28)
start_date = end_date - timedelta(days=730)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
work_df = pd.DataFrame({'snapshot_datetime': date_range})
def is_workday(date):
    if date.weekday() >= 5 or date in holiday_dates:
        return 0
    return 1
work_df['workday'] = work_df['snapshot_datetime'].apply(is_workday)

print(work_df)
```
